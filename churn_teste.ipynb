Análise exploratória para avaliar o perfil de churn de um banco

```python
import pandas as pd

df_churn = pd.read_csv('D:\churn.csv')
df_churn.head()
```

```python
#retirar colunas: customerid e surname pq são unicas e nao agregam valor. geralmente as identificadoras. numericas qualitativas
#.drop faz deletarr
# parametro axis = 1 diz q é pra deletar colunas, axis=0 é pra linha
#parametro inplace = True faz substituir no df original a retirada do drop
array_delete = ['RowNumber','CustomerId', 'Surname']
df_churn.drop(array_delete, axis=1, inplace = True)
df_churn.head()
```

```python
#.loc = filtra na tabela o segmento inserido
# query e filter tbm sao funções que filtra
df_churn.loc[(df_churn.Age == 30) & (df_churn.Tenure == 3)].head(10)
```

```python
#.iloc = me retorna toda a linha inserida
df_churn.iloc[4]
```

```python
#Me retorna uma previa das colunas
df_churn.info()
```

```python
#separar as features(colunas) qualitativas das quantitativas
# na quanti fazer um estudo estatistico e nas quali fazer um estudo de distribuição classe
#como o exited é a variavel alvo, colocar nas duas arrays


array_quali = ['Geography', 'Gender', 'HasCrCard','IsActiveMember','Exited']
array_quanti = ['CreditScore','Tenure','Age','Balance','NumOfProducts','EstimatedSalary','Exited']

df_churn_quali = df_churn[array_quali]
df_churn_quanti = df_churn[array_quanti]
```

```python
df_churn_quali
```

```python
df_churn_quanti.describe().round(2)

# Tenure é o nível de estabilidade do cliente em possuir um salario. 0-10
# Balance = Saldo Bancário 
# Estimated Salary é anual 
```

```python
#value_counts = me retorna os valores distintos da coluna


df_churn['Geography'].value_counts() 
```

```python
# com parametro (normalize) me retorna a % 
df_churn['Age'].value_counts(normalize = True).head(10) * 100
```
